```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(12, 24))

for sp in range(0,6):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')


for i in range(5):
    ax = fig.add_subplot(6, 3, i+7)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[i]], linewidth=3, c=cb_dark_blue)
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[i]], linewidth=3, c=cb_orange)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[i])
    ax.tick_params(top='off', bottom='off', left='off', right='off', labelbottom='off')
    
    if i == 0:
        ax.text(1970, 80, 'Women')
        ax.text(1970, 15, 'Men')
    elif i == 4:
        ax.text(1970, 30, "Women")
        ax.text(1970, 68, "Men")


for j in range(0,6):
    ax = fig.add_subplot(6,3,j+12)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[j]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[j]], c=cb_orange, linewidth=3)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(other_cats[j])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
    if j+12 == 15 or j+12 == 16 or j+12 == 17:
        ax.tick_params(labelbottom='on')
    if j == 0:
        ax.text(2002, 90, 'Women')
        ax.text(2002, 6, 'Men')
    elif j == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
plt.show()
```
